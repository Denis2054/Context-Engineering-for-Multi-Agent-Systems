## ⚡ Flash Attention 2: The Triton Kernel for Attention

Standard attention computes the full (seq_len × seq_len) score matrix
and writes it to HBM before computing the softmax. At 131K tokens this
is a catastrophic memory cost:

    131,072 × 131,072 × 2 bytes (bfloat16) = ~34 GB
    just for a single attention layer's intermediate matrix.

Flash Attention 2 eliminates this entirely using the same architectural
philosophy as the Triton Planner Kernel:

| Triton Planner Kernel       | Flash Attention 2              |
|-----------------------------|--------------------------------|
| Load events into SRAM once  | Load Q,K,V tiles into SRAM     |
| Atomic operations in-register| Softmax computed in-register  |
| Single write-back to HBM    | Single write-back to HBM       |
| No intermediate materialization | No score matrix in HBM    |

The result is that attention memory cost drops from O(seq_len²) to
O(seq_len) — the quadratic wall that makes long-context inference
prohibitively expensive on standard kernels is removed entirely.

On the H100 with bfloat16, Flash Attention 2 typically delivers
2-4x faster attention computation and uses 5-20x less VRAM for
long sequences compared to the HuggingFace default path.

This is not an approximation. The outputs are mathematically
identical to standard attention — only the execution path changes.
The DeepSeek weights and reasoning behavior are completely unaffected.

```python
import subprocess
import sys
import importlib

def is_flash_attn_installed():
    try:
        import flash_attn
        return True
    except ImportError:
        return False

if is_flash_attn_installed():
    import flash_attn
    print(f"✅ Flash Attention already installed: v{flash_attn.__version__}")
    print(f"   Skipping installation — session cache active.")
else:
    print("⚙️  Flash Attention not found. Installing...")
    subprocess.run(
        [sys.executable, "-m", "pip", "install",
         "flash-attn", "--no-build-isolation"],
        check=True
    )
    print("✅ Flash Attention installed successfully.")

# Final verification
import flash_attn
print(f"⚡ Flash Attention v{flash_attn.__version__} ready.")
```

The inference cell is the exact equivalent: everything else in your DeepSeek notebook — mounting Drive, loading weights, tokenizing — is a one-time or negligible cost.

What Each Change Does
torch.backends.cuda.enable_flash_sdp(True) forces the attention computation through Flash Attention, which is structurally identical to your fused kernel philosophy — it loads Q, K, V into SRAM once and never writes intermediate attention scores back to HBM.

On H100 with bfloat16 at your context length, this eliminates the dominant memory bandwidth cost.

torch.compile with inductor backend is the practical equivalent of what you did manually in Triton for the planner. Instead of writing the kernel yourself, Inductor analyzes the 32-layer forward pass and generates fused Triton kernels automatically — collapsing operations that HuggingFace executes as separate CUDA calls into single kernel launches.
torch.inference_mode() is a harder version of torch.no_grad(). It disables the autograd graph entirely, freeing both memory and compute that gradient tracking consumes even during inference.

The compilation step will take 60-120 seconds the first time — this is the JIT cost equivalent to Triton's warm-up in your planner. After that first run, the compiled kernels are cached and subsequent calls pay only the inference cost.what cell does it replace?17:29It replaces this cell.

The Key Addition: attn_implementation="flash_attention_2"
This is the critical line. It tells HuggingFace to wire Flash Attention 2 directly into the model architecture at load time, rather than injecting it later. This means by the time your inference cell runs, the attention mechanism is already pointing at the optimized kernel. The two cells work together as a pipeline:    

| **Load cell**                 | **Inference cell**                     |
|------------------------------|-----------------------------------------|
| `attn_implementation=FA2`    | `torch.compile` locks it in             |
| `device_map={"": 0}`         | no device transfers during `generate()` |
| `model.eval()`               | `torch.compile(fullgraph=True)` cleaner |
| bfloat16 explicit            | tensor cores fully utilized             |


```python
# ============================================================
# CELL: Triton-Accelerated Inference
# Replaces the standard model.generate() cell
# DeepSeek weights unchanged — only the execution engine upgrades
# ============================================================

import time
import torch
import triton
from torch.nn.attention import SDPBackend, sdpa_kernel

if install_deepseek == False:

    # --- STEP 1: Inject Triton Flash Attention into the loaded model ---
    # This replaces HuggingFace's default SDPA with a Triton-optimized kernel
    # DeepSeek weights are untouched — only the compute path changes

    # Force H100 to use Flash Attention backend (Triton-compiled path)
    torch.backends.cuda.enable_flash_sdp(True)
    torch.backends.cuda.enable_math_sdp(False)
    torch.backends.cuda.enable_mem_efficient_sdp(False)

    # --- STEP 2: Compile ONCE per session ---
    # torch.compile wraps the model object in-place.
    # Re-wrapping an already-compiled model causes errors and
    # doubles the JIT cost. The flag prevents this.
    if not getattr(model, '_triton_compiled', False):
        print("🔥 Compiling model with Triton backend (one-time JIT)...")
        compile_start = time.time()

        model = torch.compile(
            model,
            backend="inductor",
            mode="reduce-overhead",
            fullgraph=False
        )

        # Brand the model object so this block never runs again
        # in this session, even if the cell is re-run
        model._triton_compiled = True

        compile_time = time.time() - compile_start
        print(f"✅ Triton compilation complete: {compile_time:.2f}s (one-time JIT cost)")
    else:
        print("✅ Triton compilation already active — skipping.")

    # --- STEP 3: Inference — runs every time ---
    print("\n🚀 Running Triton-accelerated inference...")
    start_time = time.time()

    inputs = tokenizer(prompt2, return_tensors='pt').to('cuda')

    with torch.inference_mode():
        with sdpa_kernel(SDPBackend.FLASH_ATTENTION):
            outputs = model.generate(
                **inputs,
                max_new_tokens=500,
                repetition_penalty=1.05,
                no_repeat_ngram_size=3,
                temperature=0.2,
                top_p=0.9,
                top_k=20,
                eos_token_id=tokenizer.eos_token_id,
                use_cache=True
            )

    generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)

    # --- STEP 4: Telemetry ---
    elapsed_time = time.time() - start_time
    tokens_generated = outputs.shape[1] - inputs['input_ids'].shape[1]
    throughput = tokens_generated / elapsed_time

    print(f"\n📊 TRITON INFERENCE AUDIT")
    print(f"-" * 45)
    print(f"⏱️  Inference Time:     {elapsed_time:.2f} seconds")
    print(f"🚀 Tokens Generated:   {tokens_generated}")
    print(f"⚡ Throughput:         {throughput:.1f} tokens/sec")
    print(f"📉 Baseline (HF):      9.75 seconds")
    print(f"📈 Speedup:            {9.75/elapsed_time:.2f}x")
    print(f"-" * 45)
```

The guard mechanism is model._triton_compiled = True — a flag branded directly onto the model object. Since the model lives in session memory, this flag survives across cell re-runs for the entire session lifetime. The moment the session resets or the model is reloaded, the flag disappears and compilation runs exactly once again automatically.

## Loading the model

Prepare the inference session. By setting `local_files_only=True`, we guarantee that the system is operating in a disconnected, sovereign mode with no external calls to Hugging Face during execution.

The Key Addition: attn_implementation="flash_attention_2"
This is the critical line. It tells HuggingFace to wire Flash Attention 2 directly into the model architecture at load time, rather than injecting it later. This means by the time your inference cell runs, the attention mechanism is already pointing at the optimized kernel. The two cells work together as a pipeline:
Load cell                    Inference cell
─────────────────────────    ──────────────────────────
attn_implementation=FA2  →   torch.compile locks it in
device_map={"": 0}       →   no device transfers during generate()
model.eval()             →   torch.compile fullgraph is cleaner
bfloat16 explicit        →   tensor cores fully utilized
One Prerequisite
Flash Attention 2 requires a separate install. Add this as a new cell immediately before the model loading cell:

```python
import time
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM

if install_deepseek == False:

    model_path = '/content/drive/MyDrive/genaisys/HuggingFaceCache/models--unsloth--DeepSeek-R1-Distill-Llama-8B/snapshots/71f34f954141d22ccdad72a2e3927dddf702c9de'

    # --- LOAD ONCE PER SESSION ---
    # Re-loading an 8B model costs ~15-20 seconds and re-allocates
    # the full VRAM footprint. The flag prevents this on cell re-runs.
    if 'model' not in dir() or not getattr(model, '_loaded', False):
        print("📂 Loading model into H100 VRAM (one-time per session)...")
        start_time = time.time()

        tokenizer = AutoTokenizer.from_pretrained(
            model_path,
            local_files_only=True
        )

        model = AutoModelForCausalLM.from_pretrained(
            model_path,
            local_files_only=True,
            torch_dtype=torch.bfloat16,
            device_map={"": 0},
            attn_implementation="flash_attention_2"
        )

        # Lock to eval mode — disables dropout, batch norm tracking
        # Required before torch.compile in the inference cell
        model.eval()

        # Brand the model object — survives cell re-runs
        # disappears only when session resets
        model._loaded = True

        end_time = time.time()
        elapsed_time = end_time - start_time

        # Telemetry
        vram_used = torch.cuda.memory_allocated(0) / 1e9
        vram_total = torch.cuda.get_device_properties(0).total_memory / 1e9

        print(f"\n📊 MODEL LOAD AUDIT")
        print(f"-" * 45)
        print(f"⏱️  Load Time:          {elapsed_time:.2f} seconds")
        print(f"🧠 VRAM Used:          {vram_used:.2f} GB / {vram_total:.2f} GB")
        print(f"⚡ Precision:          bfloat16 (H100 tensor cores)")
        print(f"🔒 Attention:          Flash Attention 2")
        print(f"📍 Device:             GPU 0 (single device, no offload)")
        print(f"-" * 45)

    else:
        vram_used = torch.cuda.memory_allocated(0) / 1e9
        vram_total = torch.cuda.get_device_properties(0).total_memory / 1e9
        print(f"✅ Model already loaded — skipping (session cache active).")
        print(f"-" * 45)
        print(f"🧠 VRAM Active:        {vram_used:.2f} GB / {vram_total:.2f} GB")
        print(f"-" * 45)
```

The guard uses 'model' not in dir() as the first check — this handles the cold start where the variable doesn't exist yet. The second check not getattr(model, '_loaded', False) handles the re-run case where model exists but points to something other than your loaded DeepSeek instance. Both conditions must fail before the load is skipped, making the guard robust against all re-run scenarios within the same session.

Execute the inference. Benchmark Note: On the NVIDIA H100, this 8B reasoning model achieves inference in about 10 seconds (approx. 9.75s) for complex multi-step tasks. This proves that open-source sovereign AI can match the responsiveness of proprietary cloud APIs.

```python
import time
if install_deepseek==False:
  # Record the start time
  start_time = time.time()


  # Tokenize the input
  inputs = tokenizer(prompt2, return_tensors='pt').to('cuda')

  # Generate output with enhanced anti-repetition settings
  outputs = model.generate(
    **inputs,
    max_new_tokens=500,
    repetition_penalty=1.05,             # Increase penalty to 1.5 or higher
    no_repeat_ngram_size=3,             # Prevent repeating n-grams of size 3
    temperature=0.2,                    # Reduce randomness slightly
    top_p=0.9,                          # Nucleus sampling for diversity
    top_k=20,                            # Limits token selection to top-k probable tokens
    eos_token_id=tokenizer.eos_token_id
  )

  # Decode and display the output
  generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)

  # Record the end time
  end_time = time.time()

  # Calculate the elapsed time
  elapsed_time = end_time - start_time

print(f"Time taken for inference: {elapsed_time:.2f} seconds")
```
